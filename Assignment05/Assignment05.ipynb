## Import packages for project 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure
```

## Input file image and transform to gray scale image

```python
file_image = 'cau.jpg'

im_color = io.imread(file_image)
im_gray  = color.rgb2gray(im_color)
```

## Define kernels for example

```python
ker = np.array([[0,0,0],[0,1,0],[0,0,0]]) 
im_conv = signal.convolve2d(im_gray, ker, boundary='symm', mode='same') 
```

## Plot the images 

```python
plt.figure(figsize=(10,10))
p1 = plt.subplot(2,2,1)
p1.set_title('color image')
plt.imshow(im_color)
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('gray image')
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

p3 = plt.subplot(2,2,3)
p3.set_title('convolution kernel')
plt.imshow(ker, cmap='gray')
plt.axis('off')

p4 = plt.subplot(2,2,4)
p4.set_title('convolution result')
plt.imshow(im_conv, cmap='gray')
plt.axis('off')

plt.show()
```

## Define kernels for computing image gradients

```python
ker_gradient = np.array([[1,0,-1],[1,0,-1],[1,0,-1]]) 
```

## Function for computing x_derivative and y_derivative and shows the result
### The y_derivative function uses transpose matrix of image

```python
x_result = x_derivative()
def x_derivative():
    x_result = signal.convolve2d(im_gray, ker_gradient, boundary='symm', mode='same') 
    plt.figure(figsize=(7,7))
    plt.imshow(x_result) 
    print("x derivative image")
    
    return x_result
```

```python
def y_derivative():
    y_result = signal.convolve2d(im_gray.T, ker_gradient, boundary='symm', mode='same') 
    plt.figure(figsize=(7,7))
    plt.imshow(y_result.T) 
    print("y derivative image")
    
    return y_result.T
y_result = y_derivative()
```

## Function for computing the magnitude of the gradient and Showing the result images

```python
def magnitude(x, y):
    absoulte = (x**2 + y**2)**0.5
    plt.figure(figsize=(7,7))
    plt.imshow(absoulte) 
    print("absoulte image")
```

```python
magnitude(x_result, y_result)
```

## Function for computing the direction of the gradient using tangent method and Showing the result of direction

```python
def calculate_angle():
    angle = np.arctan2(y_result, x_result)
    plt.figure(figsize=(7,7))
    plt.imshow(angle) 
    print("direction image")
```

```python
calculate_angle()
```

## Define kernels for smoothing image

```python
ker_soomth = np.array([[1/3,1/3,1/3],[1/3,1/3,1/3],[1/3,1/3,1/3]])
```

## Define function for smoothing and show the result of smoothing image function

```python
def smooth():
    s_result = signal.convolve2d(im_gray, ker_soomth, boundary='symm', mode='same') 
    
    plt.figure(figsize=(20,20))
    p5 = plt.subplot(1,2,1)
    p5.set_title('gray image')
    plt.imshow(im_gray, cmap='gray')
    plt.axis('off')
    
    p6 = plt.subplot(1,2,2)
    p6.set_title('smooth image')
    plt.imshow(s_result, cmap='gray')
    plt.axis('off')
    
    plt.show()
```

```python
smooth()
```
